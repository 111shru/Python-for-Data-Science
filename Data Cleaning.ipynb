# Regex 

```python
import re
```

```python
# regex shortform is re
```

```python
# findall return all matches in a particular string
```

```python
str1= '101:python 102:java 103:react 104:nodejs'
pattern1 = '\d'
pattern2 = '\d+'
result1= re.findall(pattern1,str1)
result2 = re.findall(pattern2,str1)
print(result1,result2)
```

```python
# d will match digit with having first digit 1
```

```python
# d+ will match with first digits legth and will match next digits with same length
```

```python
#search method
```

```python
# return match and firsy occurance of match
```

```python
result1= re.search(pattern1,str1)
result2 = re.search(pattern2,str1)
print(result1,result2)
```

```python
# span shows length and index of match
```

```python
# split function splits the string in accordance with pattern
```

```python
str1= '101:python 102:java 103:react 104:nodejs'

pattern2 = '\d+'

result2 = re.split(pattern2,str1)
print(result2)
```

```python
str1= '101:python 102:java 103:react 104:nodejs'
pattern1 = '\d'

result1= re.split(pattern1,str1)

print(result1)
```

```python
#sub method is used to substitute the pattern with given i/p
```

```python
str1= '101:python 102:java 103:react 104:nodejs'
pattern1 = '\d'

result1= re.sub(pattern1,'substitute',str1)

print(result1)
```

```python
str1= '101:python 102:java 103:react 104:nodejs'

pattern2 = '\d+'

result2 = re.sub(pattern2,'substitute',str1)
print(result2)
```

```python
# match will match for first occuring for corresponding pattern
```

```python
str1= '101:python 102:java 103:react 104:nodejs'

pattern2 = 'python'

result2 = re.match(pattern2,str1)
print(result2)
```

```python
#it will return null because match checksonly begining of str1
```

```python
str1= '101:python 102:java 103:react 104:nodejs'

pattern2 = '101'

result2 = re.match(pattern2,str1)
print(result2)
```

```python
# here result is obtained as 101 is at beginning of string
```

```python
str1= '101:python 102:java 103:react 104:nodejs'

pattern2 = 'python'
pattern3 = re.compile('101')
result2 = re.match(pattern3,str1)
print(result2)
```

# apply functions

```python
import pandas as pd
```

```python
df= pd.read_csv('train.csv', index_col='PassengerId')
df.head(10)
```

```python
def convert_to_category(x):
    if x<18:
         return'children'
    elif x>= 18 or x<50:
        return'adult'
    else:
        return'old'
    
```

```python
df['Age']= df['Age'].apply(convert_to_category)
```

```python
df
```

```python
#here Age column is applied with the function convert to category
```
