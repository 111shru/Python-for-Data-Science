# 3. Line Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('imdb_1000.csv')
```

```python
df
```

```python
plot= df[df.genre.isin(['Crime','Horror'])]
```

```python
plot
```

```python
sns.set_style('dark')
```

```python
sns.lineplot(plot.genre, plot.star_rating)
```

```python
sns.lineplot(plot.duration, plot.star_rating)
```

```python
plot.head()
```

```python
sns.lineplot(plot.star_rating, plot.genre)
```

```python
sns.lineplot(plot.star_rating, plot.duration)
```

```python
# another way to write code for lineplot is by specifying x= , y= and data=
```

```python
sns.lineplot(x='star_rating', y= 'duration', data= plot)
```

```python
#here we got the same output
```

# Hue 

```python
sns.lineplot(x='star_rating', y= 'duration', data= plot , hue= 'genre')
```

```python
# here we added annother  paramter genre to star_rating and duration, hue is use to color the 3rd aditionsl varisble and differentiate the plot
```

```python
sns.lineplot(x='star_rating', y= 'duration', data= plot , hue= 'genre', palette= ['gray','xkcd:brick red'])
```

```python
# now here we specified colors for 3rd paramter
```

```python
sns.lineplot(x='star_rating', y= 'duration', data= plot , style= 'genre')
```

```python
#here 3rd paramter is differentiatiated using style of line
```

```python
sns.lineplot(x='star_rating', y= 'duration', data= plot , hue= 'genre', style= 'genre')
```

```python
#here 3rd paramter is differentiatiated using style of line and hue
```

```python
sns.lineplot(x='star_rating', y= 'duration', data= plot , size= 'genre')
```

```python
#here 3rd paramter is differentiatiated using size
```

```python
sns.lineplot(x='star_rating', y= 'content_rating', data= plot , size= 'genre')
```

```python
# here ci means interval
```

```python
sns.lineplot(x='star_rating', y= 'content_rating', data= plot , ci= None, size= 'genre')
```

```python
sns.lineplot(x='star_rating', y= 'content_rating', data= plot , ci= None, lw= 4 , color= 'red', alpha= 0.5)
```

```python
# here alpha means transparency, lw means line width
```

```python
sns.lineplot(x='star_rating', y= 'content_rating', data= plot , ci= None, lw= 4 , color= '#aa00aa', alpha= 1)
```
