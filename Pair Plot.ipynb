```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('mtcars.csv')
```

```python
df.head()
```

```python
df.plot.scatter(x= 'hp', y= 'mpg')
```

```python
sns.set_style('white')
sns.set_style('ticks')
```

```python
sns.regplot(x= 'hp', y='wt', data=df)
```

```python
# we got scatter plot with regression line
```

```python
sns.regplot(x= 'hp', y='wt', data=df, color='r')
```

```python
sns.lmplot(x= 'hp', y='wt', data=df, hue='gear')
```

```python
sns.lmplot(x= 'hp', y='wt', data=df, hue='gear', palette='spring')
```

```python
sns.lmplot(x= 'hp', y='wt', data=df, hue='gear', palette='Set1', markers='+')
```
