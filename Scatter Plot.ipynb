# 2. Scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
```


# Simple scatter plot

```python
y1 = np.random.randn(70)
y2=  np.random.randn(70)

# both y1 and y2 are of same size

plt.scatter(y1,y2)
plt.show()
```

# Adjust axis (incase reqd)

```python
y1 = np.random.randn(70)
y2=  np.random.randn(70)

# both y1 and y2 are of same size

plt.scatter(y1,y2)
plt.axis([-3,3,-3,3])
plt.show()
```

```python
# here we got the scatter lot with predefined axes x= 3 to 3, y= 3 to 3
```

# Speicfy a common X axis and assign seperate markers for both

```python
x= np.arange(70)
y1 = np.random.randn(70)
y2=  np.random.randn(70)

# both y1 and y2 are of same size

plt.scatter(x,y1, marker ='o', label= 'y1 with x')
plt.scatter(x,y2, marker ='v', label= 'y2 with x')
plt.legend(loc= 'upper left')
plt.show()
```

```python
x= np.arange(70)
y1 = np.random.randn(70)
y2=  np.random.randn(70)

# both y1 and y2 are of same size

plt.scatter(x,y1, marker ='o', c= 'g')
plt.scatter(x,y2, marker ='v', c= 'r')
plt.legend(loc= 'upper left')
plt.show()
# here we are changing colors for markers
```
